```python
# read json file

import pandas,json,os
rootPath = '/content/drive/MyDrive/UCLA REU 2021 KG /'
def fullPath(filename):
  return os.path.join(os.path.dirname(rootPath), filename) # rootPath+filename

# importing modlue
from pandas import *
  
# reading CSV file
data = read_csv(fullPath('Topic Modeling/NMF/keywords/topic_keywords_alldata.csv'))
  
# converting column data to list
topic_keywords = {}
for i in range(100):
    topic_keywords[i] = data[f'Topic {i}'].tolist()
print(topic_keywords[99])
```

```python
import numpy as np
counter = 0
for tweet in jsons[600000:]:
  if (counter == 13949): 
    with open(f'/content/drive/MyDrive/UCLA REU 2021 KG /Embeddings/wordEmbeddings/all_tweet_with_nmf_keywords_last.json', 'w') as fp:
      json.dump(jsons[600000:], fp)
      
  tweet["keywords"] = dict()
  for topic in tweet['nmf_topics']:
    l = dict()
    for word in tweet['cleaned_text']:
      for keyword in topic_keywords[topic]:
        try:
          l[(word,keyword)] = model3.similarity(word, keyword)
        except KeyError:
          try:
            l[(word,keyword)] = model1.similarity(word, keyword)
          except:
            pass
         
    try:
      word, keyword = sorted(l, key = lambda x: l[x],reverse=True)[0]
      #print(l)
      if l[(word,keyword)] < 0.7:
        candidate = (word, keyword,l[(word,keyword)])
        raise IndexError
      if l[(word,keyword)] > 0.9999:
        tweet["keywords"][topic]=[keyword,str(1.0)]
        #print(f"topic {topic}, keyword {keyword}, word {word}, similarity 1")

      else:
        tweet["keywords"][topic]=(keyword,str(l[(word,keyword)]))
        #print(f"topic {topic}, keyword {keyword}, word {word}, similarity {l[(word,keyword)]}")

    except IndexError:
      try:
        l2 = {}
        for word in tweet['cleaned_text']:
          for keyword in topic_keywords[topic]:
            try:
              l2[(word,keyword)] = model1.similarity(word, keyword)
            except KeyError:
              pass
        word, keyword = sorted(l2, key = lambda x: l2[x],reverse=True)[0]

      except IndexError:
        #print('word not found')
        break
      if l2[(word, keyword)] > l[(candidate[0],candidate[1])]:
        if (l2[(word,keyword)] > 0.9999):
          tweet["keywords"][topic]=(keyword,str(1.0))
          #print(f"topic {topic}, keyword {keyword}, word {word}, similarity 1")
        else:
          #print(f"topic {topic}, keyword {keyword}, word {word}, similarity {l2[(word,keyword)]}")
          tweet["keywords"][topic]=(keyword,str(l2[(word,keyword)]))
      else:
        #print(f"topic {topic}, keyword {candidate[0]}, word {candidate[1]}, similarity {l[(candidate[0],candidate[1])]}")
        tweet["keywords"][topic]=(candidate[1], str(l[(candidate[0],candidate[1])]))
  counter+=1

        
```
